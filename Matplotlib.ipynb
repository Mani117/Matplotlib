```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,51, 10)
y = np.square(x) + 400*np.random.rand(6)
```

The main idea behind using the more formal object-oriented method is to create figure objects and then just call methods or attributes off of that object. This approach helps better in dealing with a canvas that has multiple plots on it. In object-oriented interface, Pyplot is used only for a few functions such as figure creation, and the user explicitly creates and keeps track of the figure and axes objects. At this level, the user uses Pyplot to create figures, and through those figures, one or more axes objects can be created. These axes objects are then used for most plotting actions.

### Adding Axes one by one to figure object

```python
## The matplotlib.figure module contains the Figure class. 
## It is a top-level container for all plot elements. 
## The Figure object is instantiated by calling the figure() 
## function from the pyplot module
## There are some parameters that can be passed to figure class.
## They are: 
## Figsize	(width,height) tuple in inches
## Dpi	Dots per inches
## Facecolor	Figure patch facecolor
## Edgecolor	Figure patch edge color
## Linewidth	Edge line width

fig = plt.figure(figsize = (12,10))
ax1 = fig.add_axes([0.7,0.1,0.5,0.5])
ax2 = fig.add_axes([0,0.1,0.5,0.5])

## add_axes has parameter rect which is a list of 4 
## values. [left, bottom, height, width].
## Left and bottom comes into picture when multiple 
## images are plotted on this figure.
##  All quantities are in fractions of figure 
## width and height.

ax1.plot(x, y, 'ys-', color = "red")

## "ys-": type of line. s denote square and - denote solid.

ax2.plot(x, x, 'go--', color = "green", label = "This is plot 2") 

## "go-": type of line. o denote circle and -- denote dashed. 
## legend(labels=['label1', 'label2', 'label3'], loc='upper right')
## There are two ways to add legend. 
## One at the time data is plotted and other is
## when through calling legend method on axes object. 
## Both shown on ax1 and ax2

ax1.legend(labels = ['This is legend 1'], loc='upper right')
ax2.legend()
ax1.set_xlabel('medium')
ax1.set_ylabel('sales')
ax1.set_title("Test data1")
ax2.set_title("Test data2")
#plt.title("Figure_Title")
plt.show()
```

### Multiplot (subplot)

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot([1,2,3])
ax2 = fig.add_subplot(221, facecolor='y')
ax2.plot([1,2,3])

## Remember it is not subplots but subplot. 
## subplot returns only one axes at a given grid location. 
## Other things are same as above.
## Subplot returns axes at fixed location. 
## Now if i want to move my plot in figure here and there, 
## i can use add_axes with rect parameter to move it.
## fig.add_subplot(rows, columns, i)  where row and columns are variable to show
## in how many parts should plot be divided and i  represent location of current 
## plot. In format used above, 
```

```python
fig = plt.figure(figsize = (12, 16))
height = 10
width = 10
row = 5
col = 4
for i in range(1, (row*col)+ 1):
    img = np.random.randint(0,255,size = (height, width), dtype = int)
    fig.add_subplot(row, col, i)
    plt.imshow(img)
    
```

### Subplots (Adding multiple plots all at once to figure object)

```python
fig,a =  plt.subplots(2,2)
x = np.arange(1,10)
a[0,0].plot(x,x*x)
a[0,0].set_title('square')
a[0,1].plot(x,np.sqrt(x))
a[0,1].set_title('square root')
a[1,0].plot(x,np.exp(x))
a[1,0].set_title('exp')
a[1,1].plot(x,np.log10(x))
a[1,1].set_title('log')

## Subplots returns a 2 dimensional array of axes. 
## These axes have same properties as shown above.
```

### Plotting different data in same axes

```python
x = np.arange(0,1001,10)
y1 = x[:30] + 100*np.random.rand(30)
y2 = x[30:60] + 150*np.random.rand(30)
y3 = x[60:] + 80*np.random.rand(41)
a = np.arange(0,51,10)
b = a + 10*np.random.rand(6)

fig = plt.figure(figsize = (8,6))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.05,0.55,0.4,0.4])
ax1.scatter(x[:30], y1, 
            color = "red", 
            alpha = 0.6, 
            marker = "o", s = 400)
ax1.scatter(x[30:60], y2, 
            color = "blue", 
            alpha = 0.5, 
            marker = "o", s = 300)
ax1.scatter(x[60:], y3, 
            color = "brown", 
            alpha = 0.9, 
            marker = "o", s = 200)
ax2.plot(a,b,"o--", color = "blue")
```
